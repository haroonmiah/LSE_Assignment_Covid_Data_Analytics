```python
# Define function to plot moving averages
def plot_moving_average(series, window, plot_intervals=False, scale=1.96):
    
    # Create a rolling window to calculate the rolling mean using the series.rolling function
    rolling_mean = series.rolling(window=window).mean()
    
    # Declare the dimensions for the plot, plot name and plot the data consisting of the rolling mean from above 
    plt.figure(figsize=(18,4))
    plt.title('Moving average\n window size = {}'.format(window))
    plt.plot(rolling_mean, 'g', label='Simple moving average trend')

    
    # Plot confidence intervals for smoothed values
    if plot_intervals:
        
        # Calculate the mean absolute square 
        mae = mean_absolute_error(series[window:], rolling_mean[window:])
        
        # Calculate the standard deviation using numpy's std function
        deviation = np.std(series[window:] - rolling_mean[window:])
        
        # Calculate the upper and lower bounds 
        lower_bound = rolling_mean - (mae + scale * deviation)
        upper_bound = rolling_mean + (mae + scale * deviation)
        
        # Name and style upper and lower bound lines and labels 
        plt.plot(upper_bound, 'r--', label='Upper bound / Lower bound')
        plt.plot(lower_bound, 'r--')
    
    # Plot the actual values for the entire timeframe
    plt.plot(series[window:], label='Actual values')
    plt.legend(loc='best')
    plt.grid(True)
```

```python
# Define function to calculate the mean absolute error
def mean_absolute_error(a, b): return abs(b - a)
```

#### Question 3.1 
- What is the difference between qualitative and quantitative data? How can these be used in business predictions?

Qualitative data is data that can be broken down into groups, categories or descriptions, e.g. treatment type, name of country or gender/sex. Quantitative data is numeric data that can be statistically manipulated, e.g., dose of treatment, populations or number of people in a cohort. 

Qualitative and quantitative data can be used in business for forecasting. Forecasting is the process of using current or past data to make informed predictions/decisions about an outcome in the future, e.g. predict the sales/profit of a business for the coming quarters based on data/trends from previous years. 

Forecasting can be further broken down to qualitative (e.g. market research) and quantitative (e.g. time-series or moving average) forecasting. When historic data is not available to make statistically relevant conclusions, you look at the qualitative method; examples include market research (getting views and opinions from a target audience) or getting expert opinions. As such, qualitative forecasting can be more expensive and time consuming. 

With qualitative forecasting, since numerical data is available to draw upon statistical conclusions, various forecasting models can be used, such as moving averages and regression modelling.
